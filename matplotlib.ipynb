```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional matplotlib
plt.plot(x,y)
plt.xlabel("X lebel")
plt.ylabel("Y lebel")
plt.title("Title")
```

```python
#subplot(row,colume,plot colume number)
plt.subplot(1,3,1)
plt.plot(x,y,'r')

plt.subplot(1,3,2)
plt.plot(y,x,'b')

plt.subplot(1,3,3)
plt.plot(y,x,'g')
```

```python
#object oriented matplotlib
fig = plt.figure()

#[left,bottom,width,heigth]
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
#fucking practice
fig1 = plt.figure()
axes=fig1.add_axes([0.1,0.1,0.8,0.8])
axes1=fig1.add_axes([0.2,0.3,0.4,0.5])
axes2=fig1.add_axes([0.3,0.5,0.1,0.2])
axes.plot(x,y,'r')
axes.set_title("Big")
axes1.plot(x,y,'b')
axes1.set_title("small")
axes2.plot(x,y,'y')
axes2.set_title("smaller")
```

```python
#part 2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#in their axes is a array of axes, we can itrate from it like
x=np.linspace(0,5,11)
y=x**2
for a in axes:
    a.plot(x,y)
```

```python
#figure size and dpi figsize(width,heigth)
fig = plt.figure(figsize=(4,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(5,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#part 3
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig = plt.figure(figsize=(4,2))
ax = fig.add_axes([0,0,1,1])
#(color,lw=linewidth,alpha=transparancy,ls=linestyle,marker=to point on the numbers)
ax.plot(x,y,color='green',lw=10,alpha=0.5,ls='--',marker="o",markersize=10,markerfacecolor='yellow',markeredgewidth=4,markeredgecolor='red')
```
